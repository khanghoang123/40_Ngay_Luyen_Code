```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

img=cv2.imread('1.jpg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()

img_gray=cv2.imread('1.jpg',0)
plt.imshow(img_gray,cmap='gray')
plt.show()

img_brightness=img.astype(np.float32)+50
img_darkness=img.astype(np.float32)-80

img_brightness=np.clip(img_brightness,0,255)
img_darkness=np.clip(img_darkness,0,255)

img_brightness=img_brightness.astype(np.uint8)
img_darkness=img_darkness.astype(np.uint8)

plt.imshow(img_brightness)
plt.show()
plt.imshow(img_darkness)
plt.show()
```
